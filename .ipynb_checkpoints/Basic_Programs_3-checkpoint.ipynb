## PHASE 3 — RECURSION (THINKING IN SELF-REFERENCE)

### Level 1: Foundation of Recursion (Base + Recursive Case)

```python
a=10
b=20
c=a+b
print(c)
```

```python
# 1. Print numbers from 1 to n using recursion.

def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)

print_numbers(5)
```

```python
# 2. Print numbers from n down to 1 using recursion.

def print_numbers(n):
    if n == 0:
        return
    print(n)
    print_numbers(n - 1)
    
print_numbers(5)
```

```python
# 3. Print only even numbers from 1 to n recursively.

def even(n):
    if n == 0:
        return
    
    if n % 2 == 0:
        print(n)   
    even(n - 1)

even(10)
```

```python
# 4. Print only odd numbers from 1 to n recursively.

def odd(n):
    if n == 0:
        return
    
    if n % 2 == 1:
        print(n)   
    odd(n - 1)

odd(10)
```

```python
# 5. Print sum of first n natural numbers recursively.

def sum_n(n):
    if n == 0:       
        return 0
    return n + sum_n(n - 1)


sum_n(5)
```

```python
# 6. Print factorial of a number recursively.

def fact(n):
    if n==0:
        return True
    return n*fact(n-1)

fact(5)
```

```python
# 7. Calculate power of a number (xⁿ) using recursion.

def power(x, n):
    if n == 0:     
        return 1
    return x * power(x, n - 1)


power(2,5)
```

```python
# 8. Find nth Fibonacci number recursively.

def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

fib(6)
```

```python
# 9. Print Fibonacci series up to n terms recursively.

def fib(n, a=0, b=1):
    if n == 0:
        return
    print(a)
    fib(n - 1, b, a + b)


fib(6)
```

```python
# 10. Find sum of digits of a number recursively.

def digit(n):
    if n==0:
        return 0
    return (n % 10) + digit(n // 10)

digit(52)
```
